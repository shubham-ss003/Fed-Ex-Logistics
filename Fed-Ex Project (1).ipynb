## Data Wrangling

```python
# Step 1: Clean and convert relevant columns to numeric
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'].str.replace(',', '').str.strip(), errors='coerce')
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'].str.replace(',', '').str.strip(), errors='coerce')
```

```python
# Fill missing 'Shipment Mode' with 'Unknown' or most frequent
df['Shipment Mode'].fillna('Unknown', inplace=True)

# Fill missing 'Vendor' with 'Unknown' or most frequent
df['Vendor'].fillna('Unknown', inplace=True)

# Fill missing 'Dosage' with 'Not Specified'
df['Dosage'].fillna('Not Specified', inplace=True)

# Fill missing 'Country' with 'Unknown' or most frequent
df['Country'].fillna('Unknown', inplace=True)
```

### Data Visualization and Story Telling

#### Chart1- Numer of Shipments by Country

```python
top_countries = df['Country'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 15 Countries by Number of Shipments')
plt.xlabel('Number of Shipments')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

#### Chart 2 - Shipment Mode Distribution

```python
plt.figure(figsize=(6, 6))
df['Shipment Mode'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Shipment Mode Distribution")
plt.ylabel('')
plt.show()
```

#### Chart 3 - Unit Price 

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Unit Price'])
plt.title("Boxplot of Unit Price")
plt.xlabel("Unit Price (USD)")
plt.show()
```

#### Chart 4 -  Chart Bar Plot

```python
top5_countries = df['Country'].value_counts().head(5).index
plt.figure(figsize=(12, 6))
sns.barplot(data=df[df['Country'].isin(top5_countries)],
            x='Country', y='Line Item Value', hue='Shipment Mode')
plt.title("Avg Line Item Value by Country and Shipment Mode")
plt.ylabel("Average Line Item Value (USD)")
plt.show()
```

#### Chart 5 

```python
pivot_table = pd.pivot_table(df, index='Fulfill Via', columns='Vendor INCO Term',
                              values='Shipment Mode', aggfunc='count', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Counts by Fulfill Via & Vendor INCO Term")
plt.xlabel("Vendor INCO Term")
plt.ylabel("Fulfill Via")
plt.show()
```
